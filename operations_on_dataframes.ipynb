```python
def custom_func_jm(row):
    if row['Steps'] > 1000:
        return row['Calorie burned']
    else:
        return 0.5 * row['Calorie burned']
```

```python
def interquartile_range(s):
    return s.quantile(0.75) - s.quantile(0.25)
```

```python
def tenth_decile_range(s):
    return s.quantile(1) - s.quantile(0.9)
```
