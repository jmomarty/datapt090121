```python
plt.figure(figsize=(20, 10))  # create a plot figure

# create the first of two panels and set current axis
plt.subplot(221) # (rows, columns, panel number)
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x));

# create the second panel and set current axis
plt.subplot(222)
plt.plot(x, np.cos(x));

# create the second panel and set current axis
plt.subplot(223)
plt.plot(x, x);
# create the second panel and set current axis
plt.subplot(224)
plt.plot(x, x**2);
```

```python
# First create a grid of plots
# ax will be an array of two Axes objects
fig, ax = plt.subplots(2)

# Call plot() method on the appropriate object
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x));
```

```python
fig = plt.figure(figsize=(5, 5))
ax = plt.axes()
x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x))

plt.show()

fig = plt.figure(figsize=(5, 5))


plt.plot(x, x, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted')

plt.show()
```

```python
fig, [ax1, ax2] = plt.subplots(1, 2)
x1 = np.linspace(0, 10, 50)
ax1.plot(x1, x1)
ax1.set_title('DROITE')
ax1.set_xlabel('ABCISSE')
ax1.set_ylabel('ORDONNEE')
ax2.plot(x1, np.sin(x1))
ax2.set_title('SINUSOIDALE')
ax2.set_xlabel('ABCISSE')
ax2.set_ylabel('ORDONNEE')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
